```python
import sys
import pandas as pd
import numpy as np
import sklearn
import matplotlib
import keras
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import seaborn as sns

```

```python
# read the csv
cleveland = pd.read_csv('heart.csv')
```

```python
# print the shape of the DataFrame, so we can see how many examples we have
print( 'Shape of DataFrame: {}'.format(cleveland.shape))
print (cleveland.loc[1])

# remove missing data (indicated with a "?")
data = cleveland[~cleveland.isin(['?'])]
data.loc[280:]
# drop rows with NaN values from DataFrame
data = data.dropna(axis=0)
data.loc[280:]

data = data.apply(pd.to_numeric)
data.dtypes
# print data characteristics, usings pandas built-in describe() function
data.describe()
```

```python
X = np.array(data.drop(['target'], 1))
y = np.array(data['target'])
mean = X.mean(axis=0)
X -= mean
std = X.std(axis=0)
X /= std

X[0]
```

```python
# create X and Y datasets for training
from sklearn import model_selection

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, stratify=y, random_state=2, test_size = 0.2)
from keras.utils.np_utils import to_categorical

Y_train = to_categorical(y_train, num_classes=None)
Y_test = to_categorical(y_test, num_classes=None)
print (Y_train.shape)
print (Y_train[:10])
X_train[3]
```

```python

from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.layers import Dropout
from keras import regularizers


# define a function to build the keras model
def create_model(hidden:int):
 
    # create model
    model = Sequential()
    model.add(Dense(hidden, input_dim=13, kernel_initializer='normal', kernel_regularizer=regularizers.l2(0.001), activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(hidden, kernel_initializer='normal', kernel_regularizer=regularizers.l2(0.001), activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(hidden, kernel_initializer='normal', kernel_regularizer=regularizers.l2(0.001), activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(2, activation='softmax'))
    # compile model
    adam = Adam(lr=0.001)
    model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
    return model

 

model5 = create_model(16)
model10 = create_model(32)
model20 = create_model(64)
print(model5.summary())

```

```python

history5=model5.fit(X_train, Y_train, validation_data=(X_test, Y_test),epochs=100, batch_size=10)
history10=model10.fit(X_train, Y_train, validation_data=(X_test, Y_test),epochs=100, batch_size=10)
history20=model20.fit(X_train, Y_train, validation_data=(X_test, Y_test),epochs=100, batch_size=10)
print(history5.history['accuracy'][-1])
print(history5.history['val_accuracy'][-1])

print(history10.history['accuracy'][-1])
print(history10.history['val_accuracy'][-1])

print(history20.history['accuracy'][-1])
print(history20.history['val_accuracy'][-1])


```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# Model accuracy
plt.plot(history5.history['accuracy'])
#plt.plot(history.history['val_acc'])
plt.plot(history10.history['accuracy'])
plt.plot(history20.history['accuracy'])
plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['hidden layer :16', 'hidden layer :32','hidden layer :64'])
plt.show()
```

```python
# convert into binary classification problem - heart disease or no heart disease
Y_train_binary = y_train.copy()
Y_test_binary = y_test.copy()

Y_train_binary[Y_train_binary > 0] = 1
Y_test_binary[Y_test_binary > 0] = 1

print(Y_train_binary[:20])
```

```python
# define a new keras model for binary classification
def create_binary_model(hidden:int):
    # create model
    model = Sequential()
    model.add(Dense(hidden, input_dim=13, kernel_initializer='normal',  kernel_regularizer=regularizers.l2(0.001),activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(hidden, kernel_initializer='normal',  kernel_regularizer=regularizers.l2(0.001),activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(hidden, kernel_initializer='normal',  kernel_regularizer=regularizers.l2(0.001),activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(1, activation='sigmoid'))
    
    # Compile model
    adam = Adam(lr=0.001)
    model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
    return model

binary_model5 = create_binary_model(16)
binary_model10 = create_binary_model(32)
binary_model20 = create_binary_model(64)


print(binary_model5.summary())
```

```python
# fit the binary model on the training data
binary5_history=binary_model5.fit(X_train, Y_train_binary, validation_data=(X_test, Y_test_binary), epochs=100, batch_size=10)
binary10_history=binary_model10.fit(X_train, Y_train_binary, validation_data=(X_test, Y_test_binary), epochs=100, batch_size=10)
binary20_history=binary_model20.fit(X_train, Y_train_binary, validation_data=(X_test, Y_test_binary), epochs=100, batch_size=10)

```

```python
df2 = pd.DataFrame([['Categorical Model', 16.0, history5.history['accuracy'][-1]], 
                    ['Binary Model', 16.0, binary5_history.history['accuracy'][-1]],
                    ['Categorical Model', 32, history10.history['accuracy'][-1]],
                    ['Binary Model', 32, binary10_history.history['accuracy'][-1]],
                    ['Categorical Model', 64, history20.history['accuracy'][-1]],
                    ['Binary Model', 64, binary20_history.history['accuracy'][-1]]
                    ]                            
                            ,
                   columns=['Model_Type', 'Hidden_Layer_Size', 'Acc'],dtype=float)

	
df2 =df2.reset_index()
df2
```
